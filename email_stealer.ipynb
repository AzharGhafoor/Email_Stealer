```python
import pandas as pd
import win32com.client
import warnings

warnings.filterwarnings("ignore")

df = pd.DataFrame(columns=['from', 'time', 'subject', 'message'])
rows_list = []

outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")
inbox = outlook.GetDefaultFolder(6)
messages = inbox.Items

for idx,m in enumerate(messages):
    body = m.body
    sender = m.SenderEmailAddress
    time = m.ReceivedTime
    subject = m.Subject
    
    data = {"from":sender, "time":time, "subject":subject, "message":body}
    rows_list.append(data)
    
path = "folder_path_of_your_choice"
df = pd.DataFrame(rows_list) 
df.to_csv(path)
print("file created")
```

```python
import win32com.client
mail_re = r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+"

outlook = win32com.client.Dispatch("Outlook.Application").GetNamespace("MAPI")

sentBox = outlook.GetDefaultFolder(5)
sender = "yourid@mail.com"
messages = sentBox.Items

for msg in messages:
    email = re.findall(mail_re, msg.To)[0]
    if(email == sender):
        msg.Delete()
        
        delBox = outlook.GetDefaultFolder(3)
        mesgs = delBox.Items
        for m in mesgs:
            email = re.findall(mail_re, m.To)[0]
            if(email == sender):
                m.Delete()
                
print("Emails send/deleted removed")
```
